## Import

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor, export_text, plot_tree
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import matplotlib.pyplot as plt
import time
label_encoder = LabelEncoder()

```

### Read data from file CSV:

```python
data = pd.read_csv('../../data/processed_data.csv', index_col=0)
data.head()
```

### Data Preprocessing:

```python
#b·ªè nh·ªØng c·ªôt kh√¥ng c·∫ßn thi·∫øt
data_processed = data[(data['la_dien_thoai_cu'] == False) & (data['gia_moi'] > 1)]
data_processed = data_processed.drop(['ten','duong_dan', 'so_luong_binh_luan','cpu','gia_cu', 'thiet_ke', 'la_dien_thoai_cu', 'loai_dien_thoai'], axis = 1)

data_processed[['mau_sac', 'he_dieu_hanh',  'hang_dien_thoai','tan_so_quet', 'loai_man_hinh','loai_pin','do_phan_giai_cam_sau', 'do_phan_giai_cam_truoc']] = data_processed[['mau_sac', 'he_dieu_hanh','hang_dien_thoai', 'tan_so_quet', 'loai_man_hinh','loai_pin','do_phan_giai_cam_sau', 'do_phan_giai_cam_truoc']].apply(LabelEncoder().fit_transform)
data_processed= data_processed.dropna()
y = data_processed['gia_moi']
y = y.squeeze()
x = data_processed.drop(['gia_moi'], axis = 1)
y_train = y[:int(y.shape[0] * 0.67)]
y_test = y[int(y.shape[0] * 0.67):]
x_train = x[:int(x.shape[0] * 0.67)]
x_test = x[int(x.shape[0] * 0.67):]

x_test
```

## Create Model

```python
model = DecisionTreeRegressor(random_state=42)
```

### Model Training:

```python
time_start = time.time()
model.fit(x_train, y_train)
time_end = time.time()
time_training = time_end - time_start
```

## Make Prediction

```python
time_start = time.time()
y_pred = model.predict(x_test)
time_end = time.time()
time_predict =  time_end - time_start 
y_pred 
```

### Model Evaluation:

```python
MEA = mean_absolute_error(y_test, y_pred)
MSE = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
exc_time = time_training + time_predict
print(f'MEA: {MEA}')
print(f'MSE: {MSE}')
print(f'R^2: {r2}') 
print(f'Execute time: {exc_time}s')
```
