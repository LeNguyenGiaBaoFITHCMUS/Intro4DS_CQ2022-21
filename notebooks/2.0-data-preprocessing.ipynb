# Th∆∞ vi·ªán

```python
import numpy as np
import pandas as pd

import ast
import re

pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

```python
data = pd.read_csv("../data/raw_data.csv")
data.head()
```

# Tr√≠ch xu·∫•t thu·ªôc t√≠nh c∆° b·∫£n

## Thu·ªôc t√≠nh: M√†u s·∫Øc, Phi√™n b·∫£n b·ªô nh·ªõ, Gi√° m·ªõi, Gi√° c≈©

```python
data["mau_sac_phien_ban_bo_nho_gia_moi_gia_cu"] = data["mau_sac_phien_ban_bo_nho_gia_moi_gia_cu"].apply(ast.literal_eval)
data = data.explode("mau_sac_phien_ban_bo_nho_gia_moi_gia_cu").reset_index(drop=True)
df_expanded = pd.DataFrame(data['mau_sac_phien_ban_bo_nho_gia_moi_gia_cu'].tolist(), columns=['mau_sac', 'phien_ban_bo_nho', 'gia_moi', 'gia_cu'])
data = pd.concat([data, df_expanded], axis=1).drop("mau_sac_phien_ban_bo_nho_gia_moi_gia_cu", axis=1)
```

## Thu·ªôc t√≠nh: M√†n h√¨nh, H·ªá ƒëi·ªÅu h√†nh, Camera sau, Camera tr∆∞·ªõc, CPU, RAM, B·ªô nh·ªõ trong, Th·∫ª SIM, Dung l∆∞·ª£ng pin, Thi·∫øt k·∫ø

```python
explode_specifications = data.thong_so_ky_thuat.apply(ast.literal_eval).apply(pd.Series)
explode_specifications.columns = ['man_hinh', 'he_dieu_hanh', 'camera_sau', 'camera_truoc', 'cpu', 'ram', 'bo_nho_trong', 'the_sim', 'dung_luong_pin', 'thiet_ke']
data = pd.concat([data, explode_specifications], axis=1)
data.drop("thong_so_ky_thuat", axis=1, inplace=True)
```

## Thu·ªôc t√≠nh: H√£ng ƒëi·ªán tho·∫°i

```python
def get_brand(s):
    if type(s)!=str:
        return None
    s = s.replace("ƒêi·ªán tho·∫°i", "")
    s = s.strip()
    res = s.split(" ")[0].strip()
    if res == "OPPO":
        return "Oppo"
    if "Bkav" in res:
        return "Bphone"
    if res == "Th√†nh":
        return "Th√†nh H∆∞ng"
    return res

data["hang_dien_thoai"] = data["ten"].apply(get_brand)
```

## Thu·ªôc t√≠nh: C√≥ ph·∫£i l√† ƒëi·ªán tho·∫°i c≈© hay kh√¥ng?

```python
def isOld(s):
    if type(s)!=str:
        return None
    s = s.lower()
    if "c≈©" in s:
        return True
    return False

data["old1"] = data["ten"].apply(isOld)
data["old2"] = data["loai_dien_thoai"].apply(isOld)

data["la_dien_thoai_cu"] = data["old1"] | data["old2"]
data.drop(columns=["old1", "old2"], inplace=True)
```

# Chuy·ªÉn ƒë·ªïi ƒë·ªãnh d·∫°ng d·ªØ li·ªáu v√† tr√≠ch xu·∫•t th√™m thu·ªôc t√≠nh

## Thu·ªôc t√≠nh: Gi√° m·ªõi, Gi√° c≈©

```python
def cast_price(s):
    if type(s)!=str:
        return np.nan
    s = s.replace(".", "")
    numbers = re.findall(r'\d+', s)
    if len(numbers) == 0:
        return np.nan
    return int(numbers[0])
data["gia_moi"] = data["gia_moi"].apply(cast_price)
data["gia_cu"] = data["gia_cu"].apply(cast_price)
```

## Thu·ªôc t√≠nh: Th·ªùi gian b·∫£o h√†nh

```python
def cast_warranty(s):
    if type(s)!=str:
        return np.nan
    numbers = re.findall(r'\d+', s)
    if len(numbers)==0:
        return np.nan
    if len(numbers)==1:
        if numbers[0]==30:
            return 1
        return int(numbers[0])
    if numbers[0]==6 and numbers[1]==12:
        return 12
    if numbers[0]==12 and numbers[1]==12:
        return 24
    return int(numbers[0])
data["thoi_gian_bao_hanh"] = data["thoi_gian_bao_hanh"].apply(cast_warranty)
```

## Thu·ªôc t√≠nh: ƒê√°nh gi√°

```python
def cast_rating(s):
    if type(s)!=str:
        return np.nan
    numbers = re.findall(r"-?\d+(?:\.\d+)?", s)
    if len(numbers)==0:
        return np.nan
    return float(numbers[0])
data["danh_gia"] = data["danh_gia"].apply(cast_rating)
```

## Thu·ªôc t√≠nh: Phi√™n b·∫£n b·ªô nh·ªõ

```python
def cast_memory_version(s):
    if type(s)!=str:
        return [np.nan, np.nan]
    numbers = re.findall(r"-?\d+(?:\.\d+)?", s)
    if len(numbers)==0:
        return [np.nan, np.nan]
    if len(numbers)==1:
        if abs(float(numbers[0]))<4:
            return [np.nan, abs(float(numbers[0])*1024)]
        return [np.nan, abs(float(numbers[0]))]
    if abs(float(numbers[1]))<4:
        return [abs(float(numbers[0])), abs(float(numbers[1])*1024)]
    return [abs(float(numbers[0])), abs(float(numbers[1]))]
data["phien_ban_bo_nho"] = data["phien_ban_bo_nho"].apply(cast_memory_version)
df_expanded = pd.DataFrame(data['phien_ban_bo_nho'].tolist(), columns=['_ram', '_bo_nho_trong'])
data = pd.concat([data, df_expanded], axis=1).drop("phien_ban_bo_nho", axis=1)
```

## Thu·ªôc t√≠nh: RAM, B·ªô nh·ªõ trong

```python
def firstNumberRAM(s):
    if type(s)!=str:
        return np.nan
    numbers = re.findall(r"-?\d+(?:\.\d+)?", s)
    if len(numbers)==0:
        return np.nan
    res = float(numbers[0])
    return res if res < 100 else res*1.0/1024

def firstNumberROM(s):
    if type(s)!=str:
        return np.nan
    numbers = re.findall(r"-?\d+(?:\.\d+)?", s)
    if len(numbers)==0:
        return np.nan
    res = float(numbers[0])
    if res < 2:
        res = res*1024

data["ram"] = data["ram"].apply(firstNumberRAM)
data["bo_nho_trong"] = data["bo_nho_trong"].apply(firstNumberROM)
data["ram"] = data["_ram"].fillna(data["ram"])
data["bo_nho_trong"] = data["_bo_nho_trong"].fillna(data["bo_nho_trong"])
data = data.drop(columns=["_ram", "_bo_nho_trong"])
```

## Thu·ªôc t√≠nh: K√≠ch th∆∞·ªõc m√†n h√¨nh, T·∫ßn s·ªë qu√©t, ƒê·ªô s√°ng m√†n h√¨nh, Lo·∫°i m√†n h√¨nh

> Bao g·ªìm tr√≠ch xu·∫•t thu·ªôc t√≠nh

```python
def extractInches(s):
    if type(s)!=str:
        return np.nan
    s = s.lower()
    # match = re.search(r'(\d+(\.\d+)?)\s*inches', s)
    match = re.search(r'(\d+[.,]?\d*)\s*inches', s)
    result = float(match.group(1).replace(",", ".")) if match else np.nan
    return result
def extractHz(s):
    if type(s)!=str:
        return np.nan
    s = s.lower()
    match = re.search(r'(\d+(\.\d+)?)\s*hz', s)
    result = float(match.group(1)) if match else np.nan
    return result
def extractNits(s):
    if type(s)!=str:
        return np.nan
    s = s.lower()
    match = re.search(r'(\d+(\.\d+)?)\s*nits', s)
    result = float(match.group(1)) if match else np.nan
    if result < 50:
        result = result*1000
    return result
def extractType(s):
    if type(s)!=str:
        return np.nan
    s = s.upper()
    if "AMOLED" in s:
        return "AMOLED"
    if "OLED" in s:
        return "OLED"
    if "LCD" in s:
        return "LCD"
    
data["kich_thuoc_man_hinh"] = data["man_hinh"].apply(extractInches)
print(data.kich_thuoc_man_hinh.isna().sum())
data["tan_so_quet"] = data["man_hinh"].apply(extractHz)
print(data.tan_so_quet.isna().sum())
data["do_sang_man_hinh"] = data["man_hinh"].apply(extractNits)
print(data.do_sang_man_hinh.isna().sum())
data["loai_man_hinh"] = data["man_hinh"].apply(extractType)
print(data.loai_man_hinh.isna().sum())
### C√≤n ƒë·ªô ph√¢n gi·∫£i ...

data = data.drop(columns=["man_hinh"])
```

## Thu·ªôc t√≠nh: S·ªë th·∫ª SIM

```python
def extractSim(s):
    if type(s)!=str:
        return np.nan
    s = s.lower()
    if "2" in s or "dual" in s or "k√©p" in s:
        return 2
    if "kh√¥ng" in s:
        return 0
    return 1
data["so_the_sim"] = data["the_sim"].apply(extractSim)
data = data.drop(columns=["the_sim"])
```

## Thu·ªôc t√≠nh: Lo·∫°i pin, Dung l∆∞·ª£ng pin, C√¥ng su·∫•t s·∫°c

> Bao g·ªìm tr√≠ch xu·∫•t thu·ªôc t√≠nh

```python
def batteryType(s):
    if type(s)!=str:
        return ""
    s = s.lower()
    if "li-po" in s:
        return "Li-Po"
    if "li-ion" in s:
        return "Li-Ion"
    if "si/c" in s:
        return "Si/C"
    return ""
def batteryCapacity(s):
    if type(s)!=str:
        return np.nan
    s = s.lower()
    match = re.search(r'(\d+(\.\d+)?)\s*mah', s)
    result = float(match.group(1)) if match else np.nan
    return result
def chargingCapacity(s):
    if type(s)!=str:
        return np.nan
    s = s.lower()
    match = re.search(r'(\d+(\.\d+)?)\s*w', s)
    result = float(match.group(1)) if match else np.nan
    return result
data["pin"] = data["dung_luong_pin"]
data["loai_pin"] = data["pin"].apply(batteryType)
data["dung_luong_pin"] = data["pin"].apply(batteryCapacity)
data["cong_suat_sac"] = data["pin"].apply(chargingCapacity)
data = data.drop(columns=["pin"])
```

## Thu·ªôc t√≠nh: ƒê·ªô ph√¢n gi·∫£i camera tr∆∞·ªõc, ƒê·ªô ph√¢n gi·∫£i camera sau

```python
def extractMP(s):
    if type(s)!=str:
        return []
    s = s.upper()
    matches = re.findall(r'(\d+(\.\d+)?)\s*MP', s)
    result = [float(match[0]) for match in matches]
    return result
data["do_phan_giai_cam_sau"] = data["camera_sau"].apply(extractMP)
data["do_phan_giai_cam_truoc"] = data["camera_truoc"].apply(extractMP)
data = data.drop(columns=["camera_sau", "camera_truoc"])
```

## Thu·ªôc t√≠nh: H·ªá ƒëi·ªÅu h√†nh

```python
def cast_OS(s):
    if type(s)!=str:
        return "Kh√°c"
    s = s.lower()
    numbers = re.findall(r'\d+', s)
    if len(numbers)==0:
        return "Kh√°c"
    if "android" in s:
        return "Android " + str(int(numbers[0]))
    if "ios" in s:
        return "iOS " + str(int(numbers[0]))
    return "Kh√°c"
data["he_dieu_hanh"] = data["he_dieu_hanh"].apply(cast_OS)
```
